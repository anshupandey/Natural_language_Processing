```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [2.6,1.5,2.9,6.4,5.8,5.2,3.7,4.9,9.4,7.6,8.2,5.1,4.2,4.9,6.3]
price= [210,180,360,590,510,580,420,430,870,710,790,590,480,410,670]
len(area),len(price)
```

```python
x = np.array(area)
y = np.array(price)
```

```python
m = 12
c = 20
LR = 0.002
for i in range(200):
    # prediction
    yhat = m*x + c
    # get the error
    e = np.mean(np.square(y-yhat))
    
    #apply gradient descent
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    m = m - LR*dedm
    c = c - LR*dedc
    print(e)
    
    #plt.scatter(x,y)
    #plt.plot(x,yhat,c='r')
    #plt.show()
```
